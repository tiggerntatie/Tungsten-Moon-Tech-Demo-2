```python
from math import *
from random import *
from PIL import Image
from PIL import ImageDraw

starQty = 10000
texRes = 2*4096


img = Image.new("RGB", (texRes,texRes), (0,0,0,))
d = ImageDraw.Draw(img)

for i in range(starQty):
  x = uniform(0,texRes)
  costheta = uniform(-1,1)
  if costheta != 0.0:
    theta = acos( costheta )
    xwidth = 2*(int(1/sin(theta)))
    y = int(texRes*theta/pi)
    color = int(pow(randint(25, 225),2))//256
    red = color + randint(0,25)
    green = color + randint(0,25)
    blue = color + randint(0,25)

    d.line(((x,y),(x+xwidth-1,y)), fill=(red,green,blue), width=2)

# img

  #x = sin(theta)*cos(phi)
  #y = sin(theta)*sin(phi)
  #z = cos(theta)
  #b = uniform(0.1,1.0)
```
