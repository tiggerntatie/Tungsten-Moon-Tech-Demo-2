```python
from math import *
from random import *
from PIL import Image
from PIL import ImageDraw, ImageColor, ImageFont
import matplotlib
#system_fonts = matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
#print(system_fonts)
font20 = ImageFont.truetype("/LiberationMono-Bold.ttf", 80)
font18 = ImageFont.truetype("/LiberationMono-Bold.ttf", 64)


texResX = 2048
texResY = 2048
wideline = 16
narrowline = 3
fullscale = 900
xc = texResX/2
yc = texResY/2
onedivx = fullscale//10
onedivy = fullscale//10

offscale = 940
hacks = 40


img = Image.new("RGB", (texResX,texResY), ImageColor.getrgb("whitesmoke"))
d = ImageDraw.Draw(img)
d.ellipse(((xc-onedivx//5, yc-onedivy//5),(xc+onedivx//5, yc+onedivy//5)), fill="black")
for divs in range(onedivx, onedivx*11, onedivx):
  d.ellipse(((xc-divs, yc-divs),(xc+divs, yc+divs)), outline="black", width=wideline if divs in (5*onedivx,10*onedivx) else narrowline)

d.line(((xc-fullscale, yc-fullscale),(xc+fullscale, yc+fullscale)), width=narrowline, fill="black")
d.line(((xc-fullscale, yc+fullscale),(xc+fullscale, yc-fullscale)), width=narrowline, fill="black")
d.line(((xc, yc-fullscale),(xc, yc+fullscale)), width=narrowline, fill="black")
d.line(((xc-fullscale, yc),(xc+fullscale, yc)), width=narrowline, fill="black")

for hack in range(hacks):
  d.line(((xc-offscale+2*hack*(offscale//hacks),yc-offscale),(xc-offscale+2*(hack+1)*(offscale//hacks),yc-offscale-3*(offscale//hacks))), width=wideline, fill="black")
  d.line(((xc-offscale+2*hack*(offscale//hacks),yc+offscale),(xc-offscale+2*(hack+1)*(offscale//hacks),yc+offscale+3*(offscale//hacks))), width=wideline, fill="black")
  d.line(((xc-offscale,yc-offscale+2*hack*(offscale//hacks)),(xc-offscale-3*(offscale//hacks),yc-offscale+2*(hack+1)*(offscale//hacks))), width=wideline, fill="black")
  d.line(((xc+offscale,yc-offscale+2*hack*(offscale//hacks)),(xc+offscale+3*(offscale//hacks),yc-offscale+2*(hack+1)*(offscale//hacks))), width=wideline, fill="black")

color = "black"
bcolor = "whitesmoke"

for (x,y) in ([xc+fullscale//2,yc],[xc-fullscale//2,yc],[xc,yc+fullscale//2],[xc,yc-fullscale//2],):
  d.rounded_rectangle(((x-70, y-40),(x+70, y+32)),4,bcolor)
  d.text((x,y), "0.5", color, font20, anchor="mm")
for (x,y) in ([xc+fullscale,yc],[xc-fullscale,yc],[xc,yc+fullscale],[xc,yc-fullscale],):
  d.rounded_rectangle(((x-70, y-40),(x+70, y+32)),4,bcolor)
  d.text((x,y), "1.0", color, font20, anchor="mm")

d.text((xc+offscale//2-100, yc+offscale-50), "NO DATA", color, font20, anchor="lm")

img = img.resize((texResX//4, texResY//4))


img
```
